```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
spark= (SparkSession.builder
        .appName('Data_Cleaning_2')
        .getOrCreate())
```

###Reading file from CSV from Volume

```python

df= ( spark.read
     .format('csv')
     .option('header','true')
     .option('delimiter',',')
     .option('timestamps','true')
     .load('/Volumes/databricks_catalog/default/default_volume1/Customer Call List_data_cleaning.csv')
     )
display(df)
#df.show(n=10, truncate=False)
```

```python
df.printSchema()
```

###standardizing the column names

```python
import re
nc_list=[]
for c in df.columns:
  nc=c.strip()
  nc=re.sub(r'[-*/]',"",nc)
  nc=re.sub(r'[^_0-9a-zA-Z\s]',"",nc)
  nc=nc.capitalize()
  nc=re.sub(r'\s+','_',nc)
  df=df.withColumnRenamed(c,nc)
display(df)
```

##Check null column count dataset

```python
cloumn_list=[ sum(col(c).isNull().cast('int')).alias(c) for c in df.columns]

df_null_counts=df.agg(*cloumn_list)
display(df_null_counts)
```

```python
display(df.describe())
```

####trim spaces,remove non_braking sapces and replace NA with proper null values

```python
na_list=['na','Na','N/a','None','UNKnOWN','unknown','Null','null','N/A','n/a','NaN','nan','Not Available','Not available','Not provided','not provided','Not','UN','-']

for c,t in df.dtypes: 
  if t=='string':
    df=df.withColumn(c,trim(col(c)))
    df=df.withColumn(c,regexp_replace(col(c),r'^\s+',' '))
    df=df.withColumn(c,when(col(c).isin(na_list),None).otherwise(col(c)))
df.show(n=20,truncate=False)
df.printSchema()
display(df)
```

```python
cloumn_list=[ sum(col(c).isNull().cast('int')).alias(c) for c in df.columns]

df_null_counts=df.agg(*cloumn_list)
display(df_null_counts)
```

###Cleaning customer id column

```python
df=df.withColumnRenamed('Customerid','Customer_id')
df= (df.withColumn('Customer_id',when(col('Customer_id').rlike('[0-9]+'),col('Customer_id').cast('long'))
                                     .otherwise(col('Customer_id').cast('String'))))
display(df)
df.printSchema()
```

###Cleaning on firstname cloumn

```python
df=df.withColumn('First_name',trim(col('First_name')))
#apply regexp to remove all non-alphabetic characters
df=df.withColumn('First_name',regexp_replace(col('First_name'),r'[^a-zA-Z]',''))

#check if the column contains any non alphabetic characters fill with null
df= (df.withColumn('First_name',when(col('First_name').rlike('[a-zA-Z]'),col('First_name'))
                                .otherwise(lit(None))))
#define data type
df=df.withColumn('First_name',col('First_name').cast('string'))
display(df)
df.printSchema()
```

####Cleaning Last Name column

```python
#trim leading and trailing spaces
df= df.withColumn('Last_name',trim(col('Last_name')))
#apply regexp to remove all non-alphabetic characters
df= df.withColumn('Last_name',regexp_replace(col('Last_name'),r'[^a-zA-Z]',''))
#check if the column contains any non alphabetic characters fill with null
df= (df.withColumn('Last_name',when(col('Last_name').rlike('[a-zA-Z]'),col('Last_name'))
                               .otherwise(lit(None))))
#define data type
df= df.withColumn('Last_name',col('Last_name').cast('string'))
display(df)
df.printSchema()
```

####Full name column creation

```python
df=df.withColumn('Full_name',trim(concat(coalesce(col('First_name'),coalesce(lit(''),col('First_name'))),lit(' '),
                                            coalesce(col('Last_name'),coalesce(lit(''),col('Last_name')))))
                         )
display(df)
df.printSchema()                    
          
```

####Email column cleanup

```python
df=df.withColumn('Email',trim(col('Email')))
df=(df.withColumn('Email',when(col('Email').rlike('[0-9a-zA-Z._]+@[0-9a-zA-Z-]+.[0-9a-zA-Z.-]'),col('Email'))
                           .otherwise(lit(None))))
df=df.withColumn('Email',col('Email').cast('string'))
display(df)
df.printSchema()
```

####Gender Column Cleanup

```python

df=df.withColumn('Gender',trim(col('Gender')))
df= (df.withColumn('Gender',when(lower(col('Gender')).isin(['m','male']),'Male')
                   .when(lower(col('Gender')).isin(['f','female']),'Female')
                   .when(lower(col('Gender')).isin(['o','other']),'Other')
                   .otherwise(None)
                   ))
df=df.withColumn('Gender',col('Gender').cast('string'))
display(df)
df.printSchema()
```

###Addres column cleanup.Creating new states,country columns

```python
from pyspark.sql.functions import col, trim, split, length, regexp_replace, when

#Address cloumn
df = df.withColumn(
    'Address',
    trim(col('Address'))
)

#Country Column
df = df.withColumn('country_clean1',when(size(split(col('Address'),','))>2,
                                            split(col('Address'), ',').getItem(2))
                                           .otherwise(
                                           split(col('Address'), ',').getItem(1)
                                          )
                          )

df=df.withColumn('country_clean2',split(trim(
                                            regexp_replace(
                                                trim(col('country_clean1')),
                                                r'[^\sa-zA-Z]','')),
                                                ' ')
                                 )
df=df.withColumn('Country',when(size(col('country_clean2'))==1,
                                                initcap(col('country_clean2').getItem(0)))
                                 .otherwise(initcap(col('country_clean2').getItem(1))
                                 )
                                ).drop(col('country_clean1'),col('country_clean2'))

display(df)
df.printSchema()
```

```python
#states Column

df= (df.withColumn('state_clean1',initcap(when (size(split(trim(col('Address')),','))>2,
                                              split(trim(col('Address')),',')[1])
                                      .otherwise(split(trim(
                                            regexp_replace(
                                                trim(split(trim(col('Address')),',')[1]),
                                                r'[^\sa-zA-Z]','')),
                                                ' ')[0])
                                               )
                        )
           )
df=df.withColumn('State',initcap(when(col('state_clean1')==col('Country'),None)
                                             .otherwise(col('state_clean1'))
                                             )
                             )

display(df)
df.printSchema()
```

###Paying_customer and Do_not_contact  column clean up

```python
df= (df.withColumn('Paying_coustomer_status',when(lower(col('Paying_customer')).isin(['yes','y']),'Yes')
                                           .when(lower(col('Paying_customer')).isin(['no','n']),'No')
                                            .otherwise(None))
                                            .drop(col('Paying_customer'))
                                        )
display(df)
df.printSchema()
```

```python
df= (df.withColumn('Do_not_contact',when(lower(col('Do_not_contact')).isin(['yes','y']),'Yes')
                                           .when(lower(col('Do_not_contact')).isin(['no','n']),'No')
                                            .otherwise(None))
                                        )
display(df)
df.printSchema()
```

####Dealing with amount cloumn creating Amount_dollars,Amount_RS

```python
df=df.withColumn('Amount',trim(regexp_replace(col('Amount'),r'[^\d$]','')))

df= (df.withColumn('Amount_dollars',when(col('country')=='India',
                                                    ceil(round((col('Amount').cast('Integer'))/85,2)))
                                              .when(col('country')=='Us',
                                                    regexp_replace(col('Amount'),r'[^\d]','').cast('Integer'))
                                              .otherwise(None)
                                              ))
df= (df.withColumn('Amount_RS',when(col('country')=='Us',
                                                    (regexp_replace(col('Amount'),r'[^\d]','')*85).cast('long'))
                                              .when(col('country')=='India',
                                                    regexp_replace(col('Amount'),r'[^\d]','').cast('long'))
                                              .otherwise(None)
                                              ))
display(df)
df.printSchema()
```

###Phone number column handeling

```python

df=df.withColumn('Phone_number',regexp_replace(col('Phone_number'),r'[^0-9]',''))
df= (df.withColumn('Phone_number',when((col('Phone_number').rlike('[0-9]')) & 
                                      (length(col('Phone_number'))==10),col('Phone_number'))
                                        .otherwise(lit(None))))
df= (df.withColumn('Phone_number',when(col('Country')=='India',
                                        concat(lit('+91 '),col('Phone_number')[0:5],lit(' '),col('Phone_number')[5:5]))
                                 .when(col('Country')=='Us',
                                concat(lit('+1-'),col('Phone_number')[0:3],lit('-'),col('Phone_number')[3:3],lit('-'),col('Phone_number')[6:4]))#[index,how much length we need]
                                .otherwise(None))
                                            )
display(df)
df.printSchema()
```

###Paid_date column cleanup

```python
df=df.withColumn('Paid_date',regexp_replace(col('Paid_date'),r'[\s-]','/'))
df=df.withColumn('Paid_date',when((col('Paid_date').isNotNull()) & (col('Paid_date').rlike(r"\d{1,2}/\d{1,2}/\d{4}")),
                                    try_to_date(col('Paid_date'),'M/d/yyyy'))
                                    .otherwise(lit(None))
                                 )
display(df)
df.printSchema()
```

#### Gmail domain column

```python
df=df.withColumn('Eamil_domain',split(split(col('Email'),'@')[1],'\\.')[0])
display(df)
df.printSchema()
```

#Drop unnecessary columns

```python
df = df.drop(
    'Not_useful_column',
    'state_clean1'
)

df = df.select(
    col("Customer_id"),
    col("First_name"),
    col("Last_name"),
    col("Full_name"),
    col("Gender"),
    col("Address"),
    col("Country"),
    col("State"),
    col("Phone_number"),
    col("Amount"),
    col("Amount_dollars"),
    col("Amount_RS"),
    col("Paid_date"),
    col("Paying_coustomer_status"),
    col("Do_not_contact"),
    col("Email"),
    col("Eamil_domain")
)
display(df)
df.printSchema()
```
